```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt  # Perbaikan nama modul dari 'matlotlib' ke 'matplotlib'

#semesta pembicaraan
x_tegangan = np.arange(0, 12, 1)
x_waktu = np.arange(0, 30, 1)
x_nk = np.arange(0, 100, 1)

#membership function
#tegangan
tegangan_buruk = fuzz.trapmf(x_tegangan, [0, 0, 2, 4])
tegangan_cukup = fuzz.trapmf(x_tegangan, [2, 4, 6, 8])
tegangan_bagus = fuzz.trapmf(x_tegangan, [6, 8, 12, 12])

#waktu
waktu_buruk = fuzz.trapmf(x_waktu, [0, 0, 5, 10])
waktu_cukup = fuzz.trapmf(x_waktu, [5, 10, 15, 20])
waktu_bagus = fuzz.trapmf(x_waktu, [15, 20, 30, 30])

#nilai kelayakan
nk_rendah = fuzz.trapmf(x_nk, [0, 0, 50, 80])
nk_tinggi = fuzz.trapmf(x_nk, [50, 80, 100, 100])

# Kurva Membership Function
fig, (ax0, ax1, ax2) = plt.subplots(nrows=3, figsize=(8, 9))  # Perbaikan jumlah subplot

ax0.plot(x_tegangan, tegangan_buruk, 'b', linewidth=1.5, label='Buruk')
ax0.plot(x_tegangan, tegangan_cukup, 'g', linewidth=1.5, label='Cukup')
ax0.plot(x_tegangan, tegangan_bagus, 'r', linewidth=1.5, label='Bagus')  # Perbaikan nama variabel
ax0.set_title('Tegangan')
ax0.legend()

ax1.plot(x_waktu, waktu_buruk, 'b', linewidth=1.5, label='Buruk')
ax1.plot(x_waktu, waktu_cukup, 'g', linewidth=1.5, label='Cukup')
ax1.plot(x_waktu, waktu_bagus, 'r', linewidth=1.5, label='Bagus')  # Perbaikan nama variabel
ax1.set_title('Waktu')
ax1.legend()

ax2.plot(x_nk, nk_rendah, 'b', linewidth=1.5, label='Rendah')
ax2.plot(x_nk, nk_tinggi, 'r', linewidth=1.5, label='Tinggi')
ax2.set_title('Nilai Kelayakan')  # Perbaikan method plot ke set_title
ax2.legend()

for ax in (ax0, ax1, ax2):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()

plt.tight_layout()
```
